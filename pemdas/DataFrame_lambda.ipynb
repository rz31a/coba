```python
import pandas as pd

data = {'Nama': ['Andi', 'Budi', 'Cici'],
        'Umur': [25, 30, 22],
        'Gaji': [5000, 6000, 4500]}

df_kamus1 = pd.DataFrame(data)

def kategori_umur(umur):
    if umur < 25:
        return 'Muda'
    elif umur >= 25:
        return 'Dewasa'

df_kamus1
 
```

```python
# apply adalah mengulang setiap baris
df_kamus1['Kategori_Umur'] = df_kamus1['Umur'].apply(kategori_umur)

df_kamus1
```

```python
# Menggunakan lambda untuk menghitung bonus (10% dari gaji)
df_kamus1['Bonus'] = df_kamus1['Gaji'].apply(lambda x: 0.1 * x)

df_kamus1
```

```python
import pandas as pd

# Membaca file Excel
file_path = "D:/@itenas/latihan web/latihan pemdas(phython)/data_sampah.xlsx"
df = pd.read_excel(file_path)

# Menampilkan DataFrame asli
print("DataFrame asli:")
print(df)

# Meminta input tahun dari pengguna
tahun_tertentu = int(input("Masukkan tahun yang diinginkan: "))

# Menghitung total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat untuk tahun tertentu
total_produksi_sampah = df[df['tahun'] == tahun_tertentu]['jumlah_produksi_sampah'].sum()
print(f"\nTotal produksi sampah di Jawa Barat untuk tahun {tahun_tertentu}: {total_produksi_sampah} ton")

# Menjumlahkan data per tahun
jumlah_per_tahun = df.groupby('tahun')['jumlah_produksi_sampah'].sum().reset_index()
print("\nJumlah produksi sampah per tahun:")
print(jumlah_per_tahun)

# Menjumlahkan data per Kota/Kabupaten per tahun
jumlah_per_kota_per_tahun = df.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_produksi_sampah'].sum().reset_index()
print("\nJumlah produksi sampah per Kota/Kabupaten per tahun:")
print(jumlah_per_kota_per_tahun)

# Export hasil akhir menjadi CSV dan Excel
jumlah_per_kota_per_tahun.to_csv('jumlah_per_kota_per_tahun.csv', index=False)
jumlah_per_kota_per_tahun.to_excel('jumlah_per_kota_per_tahun.xlsx', index=False)
print("\nData telah diexport menjadi file CSV dan Excel.")

# Menggunakan iterrows
for index, row in jumlah_per_kota_per_tahun.iterrows():
    print(f"Kabupaten/Kota: {row['nama_kabupaten_kota']}, Tahun: {row['tahun']}, Jumlah Produksi Sampah: {row['jumlah_produksi_sampah']} ton")

# Langkah tambahan seperti membuat video penjelasan, meng-upload ke GitHub, dan mengirim PDF ke e-learning biasanya dilakukan secara manual.
```

```python
import pandas as pd

# Membaca file Excel
file_path = "D:/@itenas/latihan web/latihan pemdas(phython)/data_sampah.xlsx"
df = pd.read_excel(file_path)
```

```python
# Menampilkan DataFrame asli
print("DataFrame asli:")
print(df)
```

```python
tahun_tertentu = int(input("Masukkan tahun yang diinginkan: "))
```

```python
# Menghitung total produksi sampah di seluruh Kabupaten/Kota di Jawa Barat untuk tahun tertentu
total_produksi_sampah = df[df['tahun'] == tahun_tertentu]['jumlah_produksi_sampah'].sum()
print(f"\nTotal produksi sampah di Jawa Barat untuk tahun {tahun_tertentu}: {total_produksi_sampah} ton")
```

```python
# Menjumlahkan data per tahun
jumlah_per_tahun = df.groupby('tahun')['jumlah_produksi_sampah'].sum().reset_index()
print("\nJumlah produksi sampah per tahun:")
print(jumlah_per_tahun)
```

```python
# Menjumlahkan data per Kota/Kabupaten per tahun
jumlah_per_kota_per_tahun = df.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_produksi_sampah'].sum().reset_index()
print("\nJumlah produksi sampah per Kota/Kabupaten per tahun:")
print(jumlah_per_kota_per_tahun)
```

```python

# Menjumlahkan data per Kota/Kabupaten per tahun dengan menggunakan iterrows
jumlah_per_kota_per_tahun = pd.DataFrame()
for index, row in df.iterrows():
    if row['tahun'] == tahun_tertentu:
        total = df[(df['nama_kabupaten_kota'] == row['nama_kabupaten_kota']) & (df['tahun'] == row['tahun'])]['jumlah_produksi_sampah'].sum()
        jumlah_per_kota_per_tahun = jumlah_per_kota_per_tahun.append({'nama_kabupaten_kota': row['nama_kabupaten_kota'], 'tahun': row['tahun'], 'jumlah_produksi_sampah': total}, ignore_index=True)

print("\nJumlah produksi sampah per Kota/Kabupaten per tahun:")
print(jumlah_per_kota_per_tahun)
```

```python
# Export hasil akhir menjadi CSV dan Excel
jumlah_per_kota_per_tahun.to_csv('jumlah_per_kota_per_tahun.csv', index=False)
jumlah_per_kota_per_tahun.to_excel('jumlah_per_kota_per_tahun.xlsx', index=False)
print("\nData telah diexport menjadi file CSV dan Excel.")
```
